```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

**Classical Not Gate == X Gate**

```python
ckt=QuantumCircuit(1)
ckt.barrier()
ckt.x(0)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```

**Classsical XOR Gate = Controlled-Z Gate**

```python
# XOR gate == C-not gate
ckt=QuantumCircuit(2)
ckt.x(0)
ckt.x(1)
ckt.barrier()
ckt.cx(0,1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```

**Classical AND Gate = Toffoli Gate**

```python
# AND gate
ckt=QuantumCircuit(3)
ckt.x(0)
ckt.x(1)
ckt.barrier()
ckt.ccx(0,1,2)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```

**Classical OR Gate = Modified Toffoli Gate**

```python
# OR gate
ckt=QuantumCircuit(3)
#ckt.x(0)
ckt.x(1)
ckt.barrier()
ckt.x(0)
ckt.x(1)
ckt.ccx(0,1,2)
ckt.x(0)
ckt.x(1)
ckt.x(2)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```

**Quantum Full-Adder (Feynman gave it !)**

```python
# Q0 and Q1 are input bits
# Q2 is input-carry bit
# Q2 gives Sum Bit
# Q3 gives ouput carry bit
# Half Adder is simply just XOR gate ...
ckt=QuantumCircuit(4)
ckt.x(0)
ckt.x(1)
ckt.x(2)
ckt.barrier()
ckt.ccx(0,1,3)
ckt.cx(0,1)
ckt.ccx(1,2,3)
ckt.cx(1,2)
ckt.cx(0,1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
```
