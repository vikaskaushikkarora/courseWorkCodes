```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

**Measuring $\psi$(-) bell state in bell basis**

```python
ckt=QuantumCircuit(2)
ckt.x(0)
ckt.x(1)
ckt.barrier()
ckt.h(0) # Create a psi(-) bell state
ckt.cx(0,1)
ckt.barrier()
ckt.cx(0,1) # Measure in bell basis
ckt.h(0)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
# If the final state after this operation is :-
# |00> = phi(+) bell state
# |01> = psi(+) bell state
# |10> = phi(-) bell state
# |11> = psi(-) bell state
# If it is superposition of these , then the initial state is mixture of more than one bell states
```

**Measuring a random state in bell basis**

```python
ckt=QuantumCircuit(2)
ckt.h(0)
ckt.barrier()
ckt.cx(1,0)
ckt.h(1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
#It gives us a superposition of all other states
```
