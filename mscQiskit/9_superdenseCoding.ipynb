```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

**Alice has to send two bits of information to Bob by transferring just one single qubit which is entangled with the qubit which Bob has**

```python
# Send an information of two bits from Alice to Bob 
# They share an entangled qubit already
# a and b are the bits of information 
a=float(input("Enter either 0 or 1 :"))
b=float(input("Enter either 0 or 1 :"))
# Create a Quantum Circuit
# Qubit 0 is for Bob
# Qubit 1 is for Alice
ckt=QuantumCircuit(2)
# Make both qubits entangled
ckt.h(0)
ckt.cx(0,1)
ckt.barrier()
# Apply some operations on qubit belonging to Alice
if (a == 1):
    ckt.x(1)
if (b == 1):
    ckt.z(1)
ckt.barrier()
# Measure in Bell Basis
ckt.cx(1,0)
ckt.h(1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
counts=result.get_counts()
print(counts)
plot_histogram(counts)
# The ouput of Q_0 is the value of a and the output of Q_1 is the value of b
```
