**Drawing and Simulating a Basic Quantum Circuit**

**1) Draw a basic Quantum Circuit**

```python
from qiskit import QuantumCircuit
qc=QuantumCircuit(2)  # Create circuit with 2 qubits
qc.h(0) # Do H-gate on q0
qc.cx(0,1) # Do CNOT on q1 controlled by q0
qc.measure_all() #Measure all qubits
qc.draw()
```

**2) Simulate a Basic Quantum Circuit**

```python
from qiskit import Aer
simulator=Aer.get_backend('aer_simulator') #You can also use qasm_simulator
counts=simulator.run(qc).result().get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
