```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
```

**X-Gate => flips qubit / reverses in Bloch-Sphere**

```python
ckt=QuantumCircuit(2)
ckt.x(0)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Z-Gate => rotates around z-axis in Bloch Sphere by $\pi$**

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Y-Gate = i$\sigma_{x}\sigma_{z}$**

```python
# y-gate == Equivalent to x-gate but gives a phase shift also
ckt=QuantumCircuit(2)
ckt.y(0)
ckt.y(1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Hadamrd Gate => Creates Superposition** 

**It is used to change between x and z basis**

**H |0> = $\dfrac{1}{\sqrt{2}}$(|0>+|1>)**

**H |1> = $\dfrac{1}{\sqrt{2}}$(|0>-|1>)**

**H $\dfrac{1}{\sqrt{2}}$(|0>+|1>) = |0>**

**H $\dfrac{1}{\sqrt{2}}$(|0>-|1>) = |1>**

```python
ckt=QuantumCircuit(2)
ckt.h(0)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Controlled-Not Gate => flips the controlled bit if the controller bit is 1**

```python
# Controlled Not-gate == flips controlled qbit if the controlloing qbit is 1 == Classical XOR gate
ckt=QuantumCircuit(2)
ckt.cx(0,1) # applied on qubit-1 , controlled by qubit-0
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Controlled-Z Gate applies Z Gate to the controlled qubit is the controller qubit is 1**

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Toffoli Gate => flips the controlled qubit if both of the controlled qubits are 1**

```python
ckt=QuantumCircuit(3)
ckt.x(0)
ckt.x(1)
ckt.ccx(0,1,2)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```

**Swap Gate simply swaps the qubits**

```python
# swap gate == swaps the input qbits
ckt=QuantumCircuit(2)
ckt.x(0)
ckt.swap(0,1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```
