**A Simple Method to Create Bell-States**

**(C-NOT)(H$\times$I)( |a>$\times$|b> ) = $\psi$(a,b)**

where :-

**$\psi$(0,0) = $\dfrac{1}{\sqrt{2}}$(|00>+|11>) = $\phi(+)$**

**$\psi$(0,1) = $\dfrac{1}{\sqrt{2}}$(|01>+|10>) = $\psi(+)$**

**$\psi$(1,0) = $\dfrac{1}{\sqrt{2}}$(|00>-|11>) = $\phi(-)$**

**$\psi$(1,1) = $\dfrac{1}{\sqrt{2}}$(|01>-|10>) = $\psi(-)$**

```python
from qiskit import *
from qiskit.tools.visualization import *
```

```python
ckt=QuantumCircuit(2)
ckt.x(1)
ckt.barrier()
ckt.h(0)
ckt.cx(0,1)
ckt.measure_all()
ckt.draw()
```

```python
simulator=Aer.get_backend('qasm_simulator')
result=execute(ckt,backend=simulator).result()
print(result.get_counts(ckt))
plot_histogram(result.get_counts(ckt))
```
