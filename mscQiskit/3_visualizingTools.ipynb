**Drawing a vector in Bloch Sphere**

```python
from qiskit import *
from qiskit.tools.visualization import *
```

**1) Single Qubit**

```python
n=1
ckt=QuantumCircuit(n)
ckt.h(0)
ckt.draw()
```

```python
simulator=Aer.get_backend('statevector_simulator')
statevector=execute(ckt,backend=simulator).result().get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
```

**2) Multiple Qubits**

```python
n=2
ckt=QuantumCircuit(n)
ckt.h(0)
ckt.x(1)
ckt.cx(0,1)
ckt.draw()
```

```python
simulator=Aer.get_backend('statevector_simulator')
statevector=execute(ckt,backend=simulator).result().get_statevector()
print(statevector)
plot_bloch_multivector(statevector)
# Please note that a bloch vector not lying on the surface represents a mixed state
#In this case , entangled state !
```

**Getting the Unitary Matrix for the Quantum Circuit**

```python
ckt=QuantumCircuit(2)
ckt.x(0)
ckt.h(1)
ckt.cx(1,0)
ckt.cz(0,1)
ckt.s(0)
ckt.draw()
```

```python
simulator=Aer.get_backend('unitary_simulator')
result=execute(ckt,backend=simulator).result()
unitary=result.get_unitary()
print(unitary)
```
