```python
import numpy as np
import matplotlib.pyplot as plt
```

**Interpolating 10 points**

```python
x=np.array([2,3,4,5,6,7,8,9,10])
y=np.array([4,23,5,6,-1,23,7,15,-5])
n=len(x)

def f(z,m):
  product=1
  for i in range(n):
    if (i != m):
      product=product*(z-x[i])/(x[m]-x[i])
  return product

def interpolf(x):
  sum=0
  for i in range(n):
    sum=sum+y[i]*f(x,i)
  return sum

z0=np.arange(-0.2+x[0],x[n-1]+0.2,0.1)
w0=interpolf(z0)
plt.plot(z0,w0,'-k')
plt.plot(x,y,'r*')
plt.grid()
plt.plot()
```
