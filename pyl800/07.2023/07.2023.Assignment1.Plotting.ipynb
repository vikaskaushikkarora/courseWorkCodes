```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import tanh
```

```python
A=1
B=1
eps1=0.1
eps2=0.01

def f(x):
  return A*tanh(x/eps1)+B*tanh(x/eps2)
```

```python
x=np.arange(-0.5,0.5,0.1)
plt.plot(x,f(x))
plt.show()
```

```python
x=np.arange(-0.5,0.5,0.01)
plt.plot(x,f(x))
plt.show()
```

```python
x=np.arange(-0.5,0.5,0.005)
plt.plot(x,f(x))
plt.show()
```
