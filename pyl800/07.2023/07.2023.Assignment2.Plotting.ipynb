```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import sin,cos
```

```python
A=1.0
B=2.0
k1=2.0
k2=3.0

def f(x):
  return A*sin(k1*x)+B*cos(k2*x)
```

```python
x=np.arange(-8,8,0.5)
plt.plot(x,f(x))
plt.grid()
plt.show()
```

```python
x=np.arange(-8,8,0.1)
plt.plot(x,f(x))
plt.grid()
plt.show()
```

```python
x=np.arange(-3,5,0.05)
plt.plot(x,f(x))
plt.grid()
plt.show()
```
