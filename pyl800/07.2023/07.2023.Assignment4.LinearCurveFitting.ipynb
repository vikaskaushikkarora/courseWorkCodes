```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data points
x=np.array([1,2,3,4,5,6,7])
y=np.array([2.1,4.3,5.7,7.9,10.2,12.5,13.8])
```

```python
n=np.size(x)
z=np.multiply(x,y)
x_square=np.multiply(x,x)
A=np.ones((2,2))
A[0,0]=sum(x)
A[0,1]=n
A[1,0]=sum(x_square)
A[1,1]=sum(x)
phi=np.array([sum(y),sum(z)])
invA=np.linalg.inv(A)
vec=np.dot(invA,phi)
m=vec[0]
c=vec[1]
ybar=m*x+c*np.ones(n)
print("m is : ",m)
print("c is : ",c)
```

```python
#plot things
plt.plot(x,y,'r*')
plt.plot(x,ybar)
plt.grid()
plt.show()
```
