```python
#import things
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data
xdata = [1,2,3,4,5,6,7,8]
ydata = [1,7,4.5,7.8,7.4,11,12.3,15.4]
xdata = np.asarray(xdata)
ydata = np.asarray(ydata)
```

```python
def fitting(xdata,ydata,n): #n is the order of fit
  fit=np.polyfit(xdata,ydata,n)
  print(fit)
  def f(x):
    sum=0
    for i in range(n+1):
      sum=sum+fit[i]*(x**(n-i))
    return sum
  plt.plot(xdata,ydata,'r*')
  x=np.arange(1,10,0.1)
  plt.plot(x,f(x))
  plt.show()
```

```python
fitting(xdata,ydata,1)
```

```python
fitting(xdata,ydata,2)
```

```python
fitting(xdata,ydata,3)
```
